##Funcion iguales:

###Parámetros:

+ *pathfas*:La ruta al fasta

###Descripción

Función que indica si la secuencia de un archivo pdb y un fasta son iguales

###Resultado

True si las secuencias son iguales

False si las secuencias no son iguales

```python
import re
from IPython.display import Image

def iguales(pathdb = str(), pathfas = str()):
    
    aa={'GLY': 'g', 'ALA': 'a', 'SER': 's', 'THR': 't'
        , 'CYS': 'c', 'VAL': 'v', 'LEU': 'l', 'ILE':'i', 
        'MET':'m','PRO':'p', 'PHE':'f', 'TYR':'y',
        'TRP':'w', 'ASP':'d', 'GLU':'e', 'ASN': 'n', 
        'GLN': 'q', 'HIS':'h', 'LYS':'k', 'ARG':'r'}
    
    db = open(pathdb, 'r')
    fas = open(pathfas, 'r')
    prot = []
    prot2 = str()d2c8sa
    endml=1
    for line in db: 
        if re.search("ATOM", line) and endml != 0:
            if re.search("CA", line):
                AA=re.findall(r'(?<=CA\s{2})\w+(?= )', line)[0]
                prot.append(AA)
        if re.search("ENDMDL", line):
            endml=0
    db.close()
    for line in fas: 
        if not re.search('>', line):
            prot2=prot2+line
    prot2=re.sub('\n', '', prot2)
    fas.close()

    prot=[aa[amin] for amin in prot]

    prot = ''.join(prot)

    return prot == prot2
```

##Conclusiones

Para encontrar similitudes entre proteínas es mucho mejor utilizar algoritmos que consideren la estructura en vez de la secuencia. Las estructuras se conservan a pesar que las secuencias divergan mucho, especialmente para proteínas lejanamente emparentadas, en otras palabras, comparar por estructura nos da más información de la similitud entre dos proteínas.
