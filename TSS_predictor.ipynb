##Parametros experimentales de entropía y entalpía

Para poder hacer el cálculo de la energía libre de Gibbs se deben tener datos experimentales por dinucleotido. Los datos fueron obtenidos de  SantaLucia J (1998) PNAS 95(4): 1460-1465.

```python
# Guardamos en un diccionario los diferentes valores de entropía y entalpía
# de las diferentes parejas de dinucleotidos 

T = 37.0
NNparams = {'AA/TT':{'H':-7.9 , 'S':-22.2} , 'AT/TA':{'H':-7.2 , 'S':-20.4},
            'TA/AT':{'H':-7.2 , 'S':-21.3},'CA/GT':{'H':-8.5,'S':-22.7}, 
            'GT/CA':{'H':-8.4 , 'S':-22.4},'CT/GA':{'H':-7.8,'S':-21.0},
            'GA/CT':{'H':-8.2 , 'S':-22.2}, 'CG/GC':{'H':-10.6 , 'S':-27.2},
            'GC/CG':{'H':-9.8 , 'S':-24.4},'GG/CC':{'H':-8.0 , 'S':-19.9},
            'G': {'H': 0.1, 'S': -2.8}, 'A': {'H': 2.3, 'S': 4.1},
            'sym': {'H': 0, 'S': -1.4}}



```

##Funciones:


### complement

####Parametros

+ *seq*: La secuencia de la cual se desea obtener la complementaria. 

####Descripción:
Obtiene la cadena complementaria de una cadena de DNA de cualquier tamaño.

####Resutlado:

Cadena complementaria



```python
def complement(seq = str()):
    #Creamos un diccionario que tiene como llave un nucleotido y como valor
    #el nucleotido con el que aparea
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'} 
    #Pasamos la cadena a una lista para poder aplicar una "Comprehension list"
    bases = list(seq)
    #Por cada base en la lista: sustituyela por su valor asociado en el diccionario    
    bases = [complement[base] for base in bases] 
    #Regresa la lista de bases como cadena
    return ''.join(bases)



```

###read 

####Dependencias:
+ **re**

####Parámetros:
+ *path* : Ruta en la cual se encuentra el archivo con las regiones a analizar

####Descripción:
Se encarga de abrir el archivo y generar un diccionario que contiene los identificadores y sus secuencias

####Resultado:

+ *seq* : Diccionario que como llave contiene el identificador del gen y como valor tiene la secuencia



```python
def read(path=str()):
    ##Abrimos el archivo con las secuencias
    arch=open(path, 'r')
    
    ## Crearemos un diccionario que tenga como llave el identificador del gen
    ## y como valor la secuencia.
    seqs = {}
    
    
    for line in arch:
        ident=re.findall(r'\w+(?= \\)', line)
        seqs[ident[0]] = ident[1]
    return seqs
```

### windowG

####Parámetros: 

+ *seq*: Un diccionario con secuencias por identificador al cual se le calcuará la energía libre de Gibbs por ventana

+ *window*: Tamaño de la ventana a partir de la cual se calculará la energía libre de Gibbs

####Descripción:

Se encarga de tomar diferentes secuencias, calcular la energía libre de Gibbs por nucleótido y regresar un diccionario con los mismos identificadores pero ahora con una lista de energías libres como valor

####Resultado:

+ *deltaG*: Diccionario que tiene como llave los identificadores de los genes y como valor una lista de energías libres por cada n



```python
def windowG (seq = dict(), window = int()):
    
    
    deltaG={}
    #pasamos un diccionario con el id como llave y la secuencia como valor
    #por cada una de las llaves:
    for key in seq.keys():
        #Se crea una llave en el diccionario deltaG que tendrá una lista con delta G para cada n de la secuencia
        deltaG[key]=[]
        #la secuencia se guarda en otra variable para facilitar el procesamiento
        secuencia = seq[key]
        tk=273.15 + T
        pro=1
        ind1= 0
        ind2= window - 1
        while(pro):
            dG=0
            if (ind2<=(len(secuencia)-1)):
                wind = secuencia[ind1:ind2+1]
                
                compwind = complement(wind)
                for i in range(0, window -1):
                    #Se crea el dinucleotido para poder acceder al diccionario NN params
                    step=wind[i:i+2]+'/'+compwind[i:i+2]
                    
                    if step not in NNparams.keys():
                        step=step[::-1]
                        
                    dG+=(((1000*NNparams[step]['H'])-
                          tk*NNparams[step]['S'])/1000)
                #Añadiendo el primer y último nucleótido.
                nt = [wind[0], wind[window-1]]
                for nucleotide in nt:
                    if nucleotide not in NNparams.keys():
                        alt=complement(nucleotide)
                        dG+=(((1000*NNparams[alt]['H'])-
                              tk*NNparams[alt]['S'])/1000)
                    else:
                         dG+=(((1000*NNparams[nucleotide]['H'])-
                              tk*NNparams[nucleotide]['S'])/1000)
                
                #Correxion por simetria
                div = window % 2
                reverse =compwind[::-1]
                if div: 
                    #Si no es divisible entonces tomamos un punto medio de referencia
                    #y no lo tomamos en cuenta en nuestra comparacion
                    sym1 = wind[0:(window/2)]
                    sym2 = reverse[0:(window/2)]
                    if sym1 == sym2:
                        #Si esl numero de ventana es divisible entre 2 entonces si
                        #no dejamos ningun nucleotido sin contar
                        dG+=(((1000*NNparams['sym']['H'])-
                              tk*NNparams['sym']['S'])/1000)
                else:
                    sym1 = wind[0:(window/2+1)]
                    sym2 = reverse[0:(window/2+1)]
                    if sym1 == sym2:
                        dG+=(((1000*NNparams['sym']['H'])-
                                tk*NNparams['sym']['S'])/1000)
                #Indexar el valor total de dG de la ventana a su respectiva lista
                deltaG[key].append(dG)
                ind1+=1
                ind2+=1
            else:
                pro = 0
        
    return deltaG
                
        
```

### predictor

####Parámetros:

+ *deltas*: Diccionario con las energías libres por cada n
+ *seq*: Diccionario de las secuencias a partir de las cuales se calcularon las energías libres
+ *cut1*: Valor de corte para D
+ *cut2*: Valor de corte para E1

####Descripción:

La función realiza el cálculo de E1, E2 y D. Para cada n evalua D y E1, si satisfacen los cortes dados se toma la región entre E1 y E2 como un TSS.

####Resultado:

+ *prediction*: Un diccionario con las secuencias de las predicciones por cada identificador
+ *E1*: Un diccionario de los valores de E1 por nt por identificador
+ *E2*: Un diccionario de los valores de E2 por nt por identificador
+ *D*: Un diccionario de los valores de D por nt por identificador
+ *positions*: Un diccionario de las posiciones a partir de las cuales se obtuvieron las predicciones.



```python
def predictor(deltas = dict(), seq = dict(), cut1 = float(), cut2=float()):
    
    #Debemos calcular D, E1 y E2 para cada uno de los n
    predictions = {}
    E1 = {}
    E2 = {}
    D = {}
    positions = {}
    #Las llaves se conservan entre deltas y seq por lo que se pueden iterar a la vez
    for key in deltas.keys():
        #Para cada llave se indica que habra una lista como valor asociado
        delG = deltas[key]#Se obtiene la secuencia que se analizara
        E1[key] = []
        E2[key] = []
        D[key] = []
        predictions[key]=[]
        pos=[]
        positions[key]=[]
        #Por cada nucleotido en la secuencia
        for n in delG:
            e2=0
            d=0
            #Se obtiene el indice del nucleotido
            n_index = delG.index(n)
            #Si aún se puede calcular E2: 
            if (len(delG)>=(n_index+199)):
            #Calculamos E1
                e1=0
                for i in range(n_index, n_index+50):
                    e1+=delG[i]
                e1=e1/50
            #Calculamos E2
                for i in range(n_index+99, n_index+199):
                    e2+=delG[i]
                e2=e2/100
            #Calculamos D y lo agregamos a la lista
                d = e1-e2
                E1[key].append(e1)
                E2[key].append(e2)
                D[key].append(d)
                
            #Si el nucleotido que se analizó tiene los valores deseados:
                if( d > cut1 and e1 > cut2):
                    sequence = seq[key]
                    #Extraemos las posiciones de la prediccion
                    #que son los 50 nucleotidos entre E1 y E2
                    predict = sequence[n_index+50:n_index+100]
                    #Comprobamos que la predicción esté a más de 25 nucleotidos
                    #de la predicción anterior
                    if pos:
                        if (((n_index+50) - pos[-1]) >=25):
                            predictions[key].append(predict)
                            positions[key].append(n_index)
                    else:
                        predictions[key].append(predict)
                        pos.append(n_index+99)
                        positions[key].append(n_index)
                    pos.append(n_index+99)
                   
                        
    return (predictions, E1, E2, D, positions)
```

##Programa principal

###Eleccion de cut-off values:

Elegimos valores de cut-off en base al artículo "Kanhere, A. & Bansal, M. (2005). A novel method for prokaryotic promoter prediction based on DNA stability. BMC Bioinformatics, 6(1):1. URL http://www.biomedcentral.com/1471-2105/6/1. En donde reportan que con valores de corte .97 y -18.37 para D y E1 respectivamente se obtiene senstividad de .9. A demás elegimos dichos valores de corte ya que de otra manera obteniamos muy pocos resultados, lo cual discutiremos más adelante.


```python
#Se leen las secuencias
seqs = read('K12_400_50_sites')
#Se calcula deltaG con una ventana de 15 nts
deltas = windowG(seqs, 15)
#Se realizan las predicciones con cut1 = .97 y cut2 = -18.37
(pred, e1, e2, d, pos) = predictor(deltas, seqs, .97, -18.37)

```

## Grafica de D, E1 y E2 promedio para cada posición n

```python


#Extraemos la longitud de la lista, vamos a truncar los calculos con respecto a la secuencia más corta.

lens=[]
for key in d.keys():
    lens.append(len(d[key]))

#Creamos listas de 0 de tamaño lon para poder indexar valores más adelante
lon = min(lens)
Dprom = [0]*lon
E1prom =[0]*lon
E2prom=[0]*lon


#Por cada posición i en cada una de las secuencias se suma su valor a su respectivo index.
for key in e1.keys():
    for i in range(0, lon):
        
        Dprom[i]+=d[key][i]
        E1prom[i]+=e1[key][i]
        E2prom[i]+=e2[key][i]

#Se obtiene el numero de identificadores, que significa lo mismo que le numero de secuencias
div = len(d.keys())
#Se obtienen los promedios
Dprom = [i/div for i in Dprom]
E1prom = [i/div for i in E1prom]
E2prom = [i/div for i in E2prom]

trace0= go.Scatter(
    x=range(len(E1prom)),
    y=E1prom,
    mode = 'lines',
    name= 'E1'
    )

trace1= go.Scatter(
    x=range(len(E2prom)),
    y=E1prom,
    mode = 'lines',
    name= 'E2'
    )

trace2= go.Scatter(
    x=range(len(Dprom)),
    y=Dprom,
    mode = 'lines',
    name= 'D'
    )
trace3 = go.Scatter(
    x=range(len(Dprom)),
    y=[.97]*len(Dprom),
    name = "D cut-off",
    line = dict(
            color = "black",
            width = 1,
            dash = "dash")
    
)

data = [trace0, trace1, trace2, trace3]
#Editando el layout

layout = dict(title = "Promedios de D, E1 y E2 para una posición n",
              xaxis = dict(title = "Posición"),
              yaxis = dict(title = "Energía libre de Gibbs") )
fig = dict(data = data, layout = layout)

py.iplot(fig, filename='Promedios de E1, E2 y D')
```



Podemos observar en la gráfica que no existe ninguna posición donde se encuentren de manera preferente los promotores, es decir, el TSS se encuentra en sitios variables a lo largo de las secuencias.

##Gráfica de las frecuencias:

Graficamos la cantidad de predicciones por ventana genómica de 10 nts para ver si había zonas enriquecidas de TSSs, en un modelo ideal esperariamos ver que la gran mayoría de las predicciones caen dentro de una misma ventana. 

```python
#Grafica de frecuencia de predicción de promotores por sitio en ventanas
#de 10 nucleotidos

max(pos.values())
freqs = {}
for i in range (10, 241, 10):
    freqs[i]=0


for i in pos.keys():
    if pos[i]:
        if (len(pos[i])>=1):
            for value in pos[i]:
                #caso especial del 0
                if value == 0:
                    freqs[10]+=1
                for key in freqs.keys():
                    if (0<=key - value <10):
                        freqs[key]+=1

                        
data = [
    go.Bar(
        x=[str(i) for i in freqs.keys()],
        y = [freqs[i] for i in freqs.keys()]
    )
]
layout = dict(title = "Frecuencia de predicción de promotor por ventana genómica",
              xaxis = dict(title = "Ventana"),
              yaxis = dict(title = "Numero de predicciones") )
fig = dict(data = data, layout = layout)

py.iplot(fig, filename = "Frecuencia de predicción de promotor por ventana genómca")
```

Claramente se ve que no existe una región donde se encuentren preferentemente los TSSs, explicando así la primera gráfica.

##Perspectivas para validar las predicciones de manera bioninformática:

Nosotros pensamos que no es posible validar éstas predicciones utilizando la función de matrix-scan de R-sat. Matrix-scan busca patrones de secuencia, lo cual depende de la identidad nucleotídica de una secuencia en un espacio limitado (los motivos de unión de TFs son de pocos nulceotidos) mientras éste predictor depende no de la identidad nucleotídica *per se* si no del contexto en el que se encuentren las secuencias. Otro argumento en contra sería que para crear la matriz que se utiliza en matrix-scan se necesitan secuencias donde se sospecha que se une el mismo TF, no hay manera de inferir eso a partir de los datos que obtuvimos.
