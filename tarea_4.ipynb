### Discrepancia entre valores DOPE y RMSD:

Los valores SCOPE que obtuvimos están asociados a un error de aproximadamente 5 Angstroms, sin embargo no hay que perder de vista que el modelo fue hecho en base a una proteína que tiene tan solo 20% de identidad con la proteína original. ### Las regiones conservadas sobrelapan:

La proteína que analizamos es perteneciente a la superfamilia de las citocromo, caracterizadas por la presencia de sólo alfa-hélices.
